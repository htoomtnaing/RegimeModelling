```python
## Version Control: Python 3.13.5

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import pathlib
from pathlib import Path
```

```python
# Define the folder path
folder_path = Path("Bloomberg_Data")
file_names = [file.name for file in folder_path.iterdir() if file.is_file()]

column_names = []

for file_name in file_names:
    # Extract the ticker prefix (assuming it's the part before the first underscore)
    ticker_prefix = file_name.split("_")[0]
    column_names.append(ticker_prefix)

dict = {}

for file_name in file_names:
    dict[file_name] = pd.read_excel(folder_path / file_name, index_col=0)
    dict[file_name].index = pd.to_datetime(dict[file_name].index)

combined = pd.concat(dict.values(), axis=1)

combined.columns = column_names
```
